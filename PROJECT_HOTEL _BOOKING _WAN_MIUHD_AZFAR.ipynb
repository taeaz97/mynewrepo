>FUNCTIONS 
---
In this **checking()** function, we can check the dataframe that we created. <br>
From shape, list of columns, data types for each column, index of the dataframe, the summary of the dataframe.<br>
So, the user choose from number 1 to number 5 according to the code to get what the user desired from the summary .<br>
This can be done so the user can understand about the dataframe before starting to explore the data. 

```python
def checking():
    while True:
        try:
            input_cond = int(input( " Choose 1:shape\n Choose 2:columns\n Choose 3:data types\n Choose 4:index\n Choose 5:describe"))
            if input_cond == 1:
                print ("These are the rows and columns of the DataFrame ",hotel_2016.shape)
            if input_cond == 2:
                print ("These are the list of columns for the DataFrame \n \n",hotel_2016.columns)
            if input_cond == 3:
                print ("These are the data types for the columns in DataFrame \n\n",hotel_2016.dtypes)
            if input_cond == 4:
                print( "The DataFrame is currently using this as index\n\n",hotel_2016.index) 
            if input_cond == 5:
                print("These are the summary of the DataFrame\n\n",hotel_2016.describe())
        except Exception as ex:
            print("Please enter number from 1 to 5")
            continue #to reset the loop by asking the input value again
        else:
            print("Yes, you got the summary!")
            break
```


These are functions without class. There are certain operation that have been put into the functions.<br>
Functions like *checking()*, *indexing()*, *drop_ren()*, are to make things easier without putting the command multiple times.<br>
The user can only key-in the function and enter the required information so that it displays the data. 

```python
# unique(), nunique()
def counting(which_column):
    counting = hotel_2016[which_column].nunique()
    what_is = hotel_2016[which_column].unique()
    print (" There are ",counting," object of ",which_column)
    print (" These are the list of ",which_column,what_is)

# creating index and resetting index
def indexing():
    input1 = int(input(" Enter 1 for set index\n Enter 2 for reset index "))
    if input1 == 1:
        input2 = str(input(" enter new index "))
        hotel_2016.set_index(input2,inplace=True)
    if input1 ==2:
        hotel_2016.reset_index(inplace=True)
    else : 
        return
    return hotel_2016

# drop and rename the dataframe
def drop_ren():
    user_in = int(input("Enter 1:drop \n Enter 2:Rename\n"))
    if user_in == 1:
        unwanted_column = str(input(" Enter column to be drop "))
        hotel_2016.drop(unwanted_column,inplace = True,axis=1)
    if user_in == 2:
        the_col = str(input(" Enter column to be renamed "))
        new_name = str(input(" Enter new name for the column "))
        hotel_2016.rename(columns={the_col : new_name}, inplace=True)

    return hotel_2016
    
```

>CLASS AND FUNCTION <br>
---
Naming the class as *table* with functions in it as methods and variables for filtering the dataframe. <br>
By using the functions we can specifically filtered desired column with specific rows that we want. <br>
Inside this class, there is also a function to display the count for choosen column using the **value_counts()** operation.

```python
class table:  ##

    def __init__ (self, user_table) :
        self.user_table = user_table
        print(self.user_table)
        
    # produce table visual    
    def table_class(self):
        self.hotel_2016 = hotel_2016
        self.user_table = self.hotel_2016
        return self.user_table
    def get_column(self,the_col):
        self.the_col = the_col
        return self.user_table[self.the_col]
    def rows_in_column(self,x_1=1,x_2=1):
        self.x_1 = x_1
        self.x_2 = x_2
        return self.user_table[self.the_col].iloc[x_1:x_2]
    #def get_info(self,list_of_column,the_object,observe):        
        #return self.user_table[self.user_table[self.the_col] == the_object][observe]
    def value_to_count(self):
        return self.user_table[self.the_col].value_counts()
    def sorting_value(self):
        return self.user_table.sort_values(by = self.the_col,ascending = False)

    def get_info(self,the_object):    
        list_observe = []
        observe = ''
        while observe != 'quit':
            observe = input( "enter column to refer to... ")
            if observe !='quit':
                list_observe.append(observe)
                return self.user_table[self.user_table[self.the_col] == the_object][[observe]]
            # try to do multiple for the column "observe" but cant work and dont know why
            if observe == 'quit':
                return self.user_table[self.user_table[self.the_col] == the_object]
            
    #function to inspect certain range of column in a table       
    def get_cord(self, x1=1,x2=1,x3=1,x4=1):
        self.x1 = x1
        self.x2 = x2
        self.x3 = x3
        self.x4 = x4
        return self.user_table.iloc[x1:x2,x3:x4]
    
    # function to compare between desired columns
    def try_column(self):
        list_of_column = []
        #self.list_of_column = list_of_column
        the_column = ''
        #self.the_column = the_column
        while the_column != 'quit':
            the_column = input("Please enter the column, or enter 'quit': ")
            if the_column != 'quit':
                list_of_column.append(the_column)
        # Show that the name has been added to the list.
        return self.user_table[list_of_column]   
```
